# ***PI 값 구하기 & 적분 값 구하기***

```python
import matplotlib.pyplot as plt
import math
import random
import numpy as np
```

## *PI값 구하기*

### 랜덤 횟수를 지정한 후 PI값과 그림 그리기.

```python
num = 0
tn = 0
d = 0
xg=np.array([])
xg2=np.array([])
yg=np.array([])
yg2=np.array([])
dpx=np.array([])
dpy=np.array([])
n1= int(input('횟수 :'))
for c in range(0,n1):
  n=random.randrange(1,1000)
  tn = tn+n

print('총 횟수 :',tn)

for i in range(0,tn):
 x = random.random()
 y = random.random()
 y1 = math.sqrt(1-x*x)
 if (y1>y):
  yin=y
  tx=np.append(xg,x)
  xg=tx
  tyin=np.append(yg,yin)
  yg=tyin
  num=num+1
  score = num/tn

 else: 
    yout=y
    tyout=np.append(yg2,yout)
    yg2=tyout
    tx2=np.append(xg2,x)
    xg2=tx2
    

num=0
PI = score*4

print('PI : ',PI)

plt.scatter(xg,yg,marker='.')
plt.scatter(xg2,yg2,marker='.')
point_x = np.arange(0,1,0.0001)
point_y = np.sqrt(1-point_x**2)
plt.plot(point_x,point_y,color='red')
plt.show()
```

### 10000번 반복을 몇 회 진행하는지 설정 후 각 횟수마다 PI값을 계산해 잔차를 구하고 그래프로 나타내기

```python
n2=int(input('횟수 : '))
n3 = 10000
n4=0
num = 0
pi2 = 3.141592
xg = np.array([])
dx = np.array([])
yg = np.array([])

for a in range(0,n2):
  for b in range(0,n3):
    x = random.random()
    y = random.random()
    y1 = math.sqrt(1-x*x)
    if (y1>y):
      num = num+1
  score = num/(n3*(a+1))
  pi=score*4
  d= pi -pi2
  pi2 = pi
  td = np.append(dx,d)
  dx= td
  n4 = n3 + n4
  tx1 = np.append(xg,n4)
  xg=tx1
  ty1 = np.append(yg,pi)
  yg=ty1
print('PI ; ',pi)
plt.plot(tx1,ty1)
plt.xlabel('count')
plt.ylabel('PI')
plt.title('PI')
plt.show()
plt.plot(tx1,dx)
plt.xlabel('count')
plt.ylabel('residual')
plt.title('Residual')
plt.show()
```

## ***적분값 구하기***


### ***랜덤 횟수 지정 후 적분값 구하고 그림 그리기***

```python
num = 0
xm = np.array([])
ym = np.array([])
tn = 0
n1 = int(input('횟수 : '))
for i in range(0,n1):
  n = random.randrange(1,1000)
  tn=n+tn
print('총 횟수 : ',tn)
for a in range(0,tn):
  x=3*random.random()
  y=2*random.random()
  y1=math.sqrt(2*x+3)-1
  if(y1>y):
    num=num+1
    tx=np.append(xm,x)
    xm=tx
    ty=np.append(ym,y)
    ym=ty
A=6*num/tn
num=0
print('면적 : ',A)
plt.scatter(tx,ty,marker='.')
point=np.arange(-1.5,4,0.01)
point_y=np.sqrt(point*2+3)-1
plt.hlines(0,-2,4,colors='black',linestyle='solid')
plt.vlines(0,-2,3,colors='black',linestyle='solid')
plt.plot(point,point_y,color='green')
plt.xlim(-2,4)
plt.ylim(-2,3)
plt.show()
```

### ***10000번 반복을 몇 회 진행하는지 설정 후 각 횟수마다 PI값을 계산해 잔차를 구하고 그래프로 나타내기***

```python
num=0
xm=np.array([])
ym=np.array([])
dx=np.array([])
n2 = int(input('횟수 : '))
n3=10000
A2=6-math.sqrt(3)
n4=n3*n2
print('총 횟수 : ',n4)
for a in range(0,n2):
  for b in range(0,n3):
    x = 3*random.random()
    y = 2*random.random()
    y1=math.sqrt(2*x+3)-1
    if(y1>y):
      num=num+1
  A=6*num/(n3*(a+1))
  n4=n4+n3
  tx=np.append(xm,n4)
  xm=tx
  ty=np.append(ym,A)
  ym=ty
  d=A-A2
  A2=A
  tdx=np.append(dx,d)
  dx=tdx
print('Area : ',A)
plt.plot(xm,ym)
plt.xlabel('count')
plt.ylabel('value')
plt.title('Area')
plt.show()
plt.plot(xm,tdx)
plt.xlabel('count')
plt.ylabel('residual')
plt.title('Residual')
plt.show()
```
